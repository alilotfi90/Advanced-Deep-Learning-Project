```python
import os
import numpy as np
from PIL import Image

parent_path = "/content/train"

# Create an empty list to store the flattened image matrices
image_matrices = []

# Use os.listdir() to get a list of all the folders in the directory
folder_list = os.listdir(parent_path)

# Loop through the folders and process each image
for folder_name in folder_list:
    # Get the path to the folder containing the images
    folder_path = os.path.join(parent_path, folder_name)

    # Create a list to store the flattened image vectors and their corresponding labels
    image_vectors = []
    image_labels = []

    # Use os.listdir() to get a list of all the items (files and folders) in the folder
    item_list = os.listdir(folder_path)

    # Loop through the items and process each file
    for item_name in item_list[:5]:  # choose the first 2 items
        # Get the path to the item
        item_path = os.path.join(folder_path, item_name)

        # Check if the item is a file (not a directory)
        if os.path.isfile(item_path):
            # Load the image file using PIL
            image = Image.open(item_path)

            # Convert the image to grayscale
            image = image.convert('L')

            # Flatten the image into a vector
            vector = np.array(image).flatten()

            # Add the flattened vector and its corresponding label to the lists
            image_vectors.append(vector)
            image_labels.append(folder_dict[folder_name])

    # Concatenate the flattened image vectors into a matrix
    image_matrix = np.column_stack(image_vectors)

    # Add the matrix and its corresponding label to the list
    image_matrices.append((image_matrix, image_labels))

# Concatenate all the image matrices into a single matrix
data_matrix = np.column_stack([matrix for matrix, label in image_matrices])

# Concatenate all the labels into a single array
label_array = np.concatenate([label for matrix, label in image_matrices])
```

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Define the number of components for the t-SNE output
n_components = 2

# Create a TSNE object with the desired number of components
tsne = TSNE(n_components=n_components)

# Fit the t-SNE model to the data matrix
tsne_result = tsne.fit_transform(data_matrix.T)

# Plot the t-SNE output
# plt.scatter(tsne_result[:,0], tsne_result[:,1], c=label_array)

# fig, ax = plt.subplots(figsize=(20, 20))
# ax.scatter(tsne_result[:,0], tsne_result[:,1], c=label_array)
# ax.text(-30, -30, "t-SNE plot of image dataset", fontsize=12)

# # Save the plot as a PNG image with 300 DPI

# plt.savefig("/content/drive/MyDrive/images/tsne_plot_Ali.png", dpi=200)
# plt.show()

fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(tsne_result[:,0], tsne_result[:,1], c=label_array)

# Add a caption to the plot
# plt.text(-30, -30, "t-SNE plot of image dataset", fontsize=12, color='black')
plt.xlabel("t-SNE plot of a subset of the image dataset")
# Save the plot as a PNG image with 300 DPI
plt.savefig("/content/drive/MyDrive/images/tsne_plot_Ali.png", dpi=300)

# Display the plot on the screen
plt.show()
```
